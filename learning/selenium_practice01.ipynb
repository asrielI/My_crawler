```python
from selenium.webdriver.common.keys import Keys
driver = webdriver.Edge(path)
driver.implicitly_wait(10)
driver.get("https://www.gamer.com.tw/")

keyword = driver.find_element_by_css_selector("input#gsc-i-id1")
keyword.send_keys("電玩")
keyword.send_keys(Keys.ENTER)

items = driver.find_elements_by_css_selector("a.gs-title")
for item in items:
    print(item.text)

driver.quit()
```

動作鏈

```python
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Edge(path)
driver.implicitly_wait(10)

url = "https://www.python.org/"
driver.get(url)

menu = driver.find_element_by_css_selector("li#about")
item = driver.find_element_by_css_selector("li#about>ul>li.tier-2.element-1")

actions = ActionChains(driver)
actions.move_to_element(menu)
actions.click(item)

actions.perform()

time.sleep(5)
driver.quit()
```
